```python
%pylab --no-import-all inline
from tensorflow.examples.tutorials.mnist import input_data
from tensorflow import __version__

digits = input_data.read_data_sets('__data', False).train
samples = np.random.randint(0, len(digits.images), 25)
digs = digits.images[samples]

fig = plt.figure()
for row in range(5):
    for col in range(5):
        plt.subplot(5,5,row*5+col+1)
        plt.imshow(np.reshape(digs[row*5+col], (28, 28)), cmap='gray')
        plt.axis('off')
# plt.subplots_adjust(wspace=-.7)

print("Congratulations, you are using TensorFlow version {0}".format(__version__))
```
