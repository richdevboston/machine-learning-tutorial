# Introduction to Linear Models

When we say linear models, we generally assume a models of the form
$$
\begin{align}
\hat{y}(\omega, x) &= \omega_0 + \sum_{i=1}^p{\omega_i x_i}
\end{align}
$$
where $\omega$ is the coefficient vector.

To illustrate the process, let us draw a sample line

```python
%pylab --no-import-all inline
from _utils import *

coeffs = (2,)
intercept = 1 # This is the 0th order coefficient
X, y = generate_linear_1D(coeffs = coeffs, intercept = intercept)
plt.scatter(X, y)
plt.title("Linear model of the form $\hat{y} = %.1f + %.1f x$"%(intercept, coeffs[0]))
plt.ylabel("$\hat{y}$")
plt.xlabel("$X$")
```

Identifying this model is fairly easy, as we can just solve a linear equation by taking two random points. However, if the input is noisy, solving the problem becomes problematic. Run the snippet below to see how a noise set might look like

```python
noise = 0.2
seed = 42
X_prime, y_prime = generate_linear_1D(coeffs = coeffs, intercept = intercept, noise = noise, seed = seed)
plt.scatter(X_prime, y_prime, label='noisy')
plt.plot(X, y, 'r', label = 'original')

plt.title("Linear model of the form $\hat{y} = %.1f + %.1f x + noise$"%(intercept, coeffs[0]))
plt.ylabel("$\hat{y}$")
plt.xlabel("$X$")
plt.legend(loc='best')
```
